```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
hbar = 1.054571817e-34 
m = 9.10938356e-31  
a = 1e-10 
```

```python
def E(n):
    return (n**2 * np.pi**2 * hbar**2) / (2 * m * a**2)

def psi(n, x):
    return np.sqrt(2/a) * np.sin(n * np.pi * x / a)

def phi(n, t):
    return np.exp(-1j * E(n) * t / hbar)

def cn(n):
    return np.sqrt(60) * (2 / (n * np.pi)**3) * (1 - (-1)**n)

def Psi(n, x, t):
    return cn(n) * psi(n, x) * phi(n, t)

def P(n, x, t):
    return np.abs(Psi(n, x, t))**2
```

```python
x_values = np.linspace(0, a, 400)
t_values = np.linspace(0, 1e-15, 3)  
```

```python
for n in range(1,6):  
    for t in t_values:
        plt.figure(figsize=(10, 6))
        

        plt.subplot(2, 1, 1)
        plt.plot(x_values, np.real(Psi(n, x_values, t)), label='Parte Real')
        plt.plot(x_values, np.imag(Psi(n, x_values, t)), label='Parte Imaginaria')
        plt.title(f"Funcin de Onda para n={n} y t={t:.2e} s")
        plt.xlabel("x")
        plt.ylabel("(x,t)")
        plt.legend()
        

        plt.subplot(2, 1, 2)
        plt.plot(x_values, P(n, x_values, t))
        plt.title(f"Distribucin de Probabilidad para n={n} y t={t:.2e} s")
        plt.xlabel("x")
        plt.ylabel("|(x,t)|^2")
        
        plt.tight_layout()
        plt.show()
```

Las funciones de onda junto a sus densidades de probabilidad asociadas se graficaron para "n" entre 1 y 5. Como se espera, debido a los coeficientes Cn encontrados analticamente, la funcin de onda se desvanece para "n" par y permanece para "n" impar.
